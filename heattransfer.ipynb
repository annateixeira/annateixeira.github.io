```python
import numpy as np
import matplotlib.pyplot as plt

Ti=40.0             #Temperatura inicial da água em graus Celsius 
Tambiente=0.0 #Temperatura ambiente em graus Celsius
T1=20.0           #Temperatura da água passando-se 10 minutos em graus Celsius
t1=10.0            #Tempo em minutos
t2=15.0           #Tempo final em minutos
k = -np.log((T1 - Tambiente)/(Ti - Tambiente))/t1

def T(t):
  return Tambiente + (Ti - Tambiente)*np.exp(-k*t) # solução da EDO
print (f'a) O coeficiente de transferencia de calor é K = {k:.1e} (1/min)'\
      .replace('.',','))
print(f'b) A temperatura da água após {t2} minutos será de {T(t2):.1f} min'\
      .replace('.',','))
t = np.linspace(0,60,10)
TT = np.linspace(Tambiente,Ti,10)
tttt, TTTT = np.meshgrid(t, TT)
ang = np.arctan( -k*(TTTT - Tambiente))
dx = np.cos(ang)
dy = np.sin(ang)
```
